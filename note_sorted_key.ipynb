sorted函数我们不陌生，是一个python内置函数，任何可迭代对象（iterable）都可以调用，返回一个排序后的新列表，不改变原对象，注意与sort区别，sort是list的一种方法。
这里我想讨论的是多级排序，也就是key怎么写。
注意：key只能是一个函数，但是可以利用元组来实现多级排序。
可以与lambda搭配使用，eg：lambda x:(A,B),这就是典型的多级排序的lambda的使用，先判断A，如果量数据的A相同，再判断B。

```python
raw_data = " post1:15:travel, post2:42:food, post3:15:tech, post4:100:food, post5:42:art"
T=[(x.split(':')) for x in raw_data.split(',')]
for x in sorted(T,key=lambda x:(-int(x[1]),x[2])):
    print(x)
```

这是一个 帖子：点赞数：标签 ，我们进行点赞数逆序排列，如果点赞数相同，则标签升序排列。
你要注意，key就是一个函数，千万不要写出来两个lambda。
然后就是一级排序里面的reverse=True，就是逆序排列，不加这个默认的是升序排列。
